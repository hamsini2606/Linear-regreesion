```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
# Step 2: Create some data (nonlinear, but we'll fit a straight line)
np.random.seed(0)
X = np.random.rand(100, 1) * 10  # X values from 0 to 10
y = 3 * X + 7 + np.random.randn(100, 1) * 5  # Linear trend with noise

```

```python
# Step 3: Fit a Linear Regression model (straight line)
model = LinearRegression()
model.fit(X, y)

```

```python
# Step 4: Predict y values and plot
y_pred = model.predict(X)

plt.scatter(X, y, color='blue', label='Data')
plt.plot(X, y_pred, color='red', label='Regression Line')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear Regression (Straight Line)')
plt.legend()
plt.grid(True)
plt.show()

```
